```python
#GROUP-BY

df = pd.read_csv('weather_data_cities.csv')

df
```

```python
#now we can group the data using their column value

g = df.groupby('city')

for city, city_df in g:
    print('City Name: ' + city)
    print('Weather Information')
    print(city_df)
    print()
```

```python
#All rows of a specific group
g.get_group('paris')
```

```python
#Find the maximum temperature and windSpeed of each city

g.max()
```

```python
#Find the average temp. and windspeed of each city

g.mean()
```

```python
#Find important data of each city

g.describe()
```

```python
#Creating Data Frame using Hash-Tables [Dictionary]

bd_wd = pd.DataFrame({
    "city" : ["Dhaka", "Sylhet", "Rajshahi"],
    "temp" : [40, 32, 44],
    "humidity" : [80, 75, 65]
})

bd_wd
```

```python
us_wd = pd.DataFrame({
    "city" : ["New York", "Chicago", "Michigan"],
    "temp" : [21, 14, 35],
    "humidity" : [68, 65, 75]
})

us_wd
```

```python
#Concate two data frame

df = pd.concat([bd_wd, us_wd])

df
```

```python
#Concate two data frame with ignoring the index will create a data frame with completely new index
df = pd.concat([bd_wd, us_wd], ignore_index=True)

df
```

```python
#Merge two data frame in a same row

df = pd.concat([bd_wd, us_wd], axis=1)
df
```

```python
#data frame that contains onlu temperature data

temp_df = pd.DataFrame({
    "city" : ["Dhaka", "Sylhet", "Rajshahi", "Chittagong"],
    "temp" : [40, 32, 44, 38],
})

temp_df
```

```python
#data frame that contains only humidity data

hum_df = pd.DataFrame({
    "city" : ["Dhaka", "Sylhet", "Chittagong"],
    "humidity" : [80, 70, 85],
})

hum_df
```

```python
#Join two df using one common column city

df = pd.merge(temp_df, hum_df, on='city')

df
```

```python
#In the previous example Rajshahi is missing as it has no value of humidity. To get is we have to join them as outer

df = pd.merge(temp_df, hum_df, on='city', how='outer')

df
```

```python
#Indexing: Pandas always take the indices from 0 to (n-1). But we can override the indices by our own value

us_wd = pd.DataFrame({
    "city" : ["New York", "Chicago", "Michigan"],
    "temp" : [21, 14, 35],
    "humidity" : [68, 65, 75]
}, index = [4, 2, 9])

us_wd
```

```python
#In pandas we can get any data from their index or their row number. If we use custom indexing the tow number may be not equal to the index number, otherwise both will be same

#Using Index number
us_wd.loc[9]
```

```python
#Using row number
us_wd.iloc[2]
```
